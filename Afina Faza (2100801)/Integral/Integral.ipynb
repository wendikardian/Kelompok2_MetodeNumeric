```python
def f(x):
    return np.sin(x)
```

```python
# Deklarasi untuk limit dari integral yang ditentukan
a = 0
b = np.pi
n = 10 

h = (b - a) / n # Untuk menghitung lebar dari setiap interval

approximation = 0 # Nilai perkiraan awals

for i in range(n): # Melakukan perulangan sebanyak jumlah interval
    approximation += f(a + i * h)# Nilai akan terus ditambahkan hingga loop berakhir

approximation *= h # Nilai perkiraan dikalikan dari lebar setiap interval

print(f"Perkiraan integral:{approximation}") # Untuk menampilkan perkiraan integral

x = np.linspace(a, b, 100)
y = f(x)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.fill_between(x, y, alpha=0.2)
plt.show()
```
