```python
# Nama  : Afina Faza
# NIM   : 2100801
# Pendidikan Ilmu Komputer A

# Eksplorasi Interpolasi Tetangga Terdekat

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Menentukan titik matriks yang akan dicari
x = np.array([0, 1, 2])
y = np.array([1, 2, 3])

x_interp = 1.5 # Inisialisasi nilai x yang akan di cari

i = np.abs(x - x_interp).argmin()# Untuk mencari nilai index i

y_interp = y[i]# Masukan nilai interpolasi y berdasarkan indeks sebelumnya

# Print the interpolated value
print(f"Nilai Interpolasi: {y_interp}") # Menampilkan Nilai Interpolasi

fig, ax = plt.subplots()
ax.scatter(x, y)
ax.scatter(x_interp, y_interp, marker='x')
plt.show()
```
