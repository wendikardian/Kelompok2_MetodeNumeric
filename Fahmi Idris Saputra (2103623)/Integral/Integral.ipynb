```python
# Fahmi Idris Saputra (2103623)
# Pendidikan Ilmu Komputer - A

# INTEGRAL

# library numpy untuk mempermudah proses perhitungan matematika
# library matplotlib untuk plotting data
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.sin(x)
```

```python
# limit hasil dari integral yang ditentukan
a = 0
b = np.pi

n = 10

# menghitung lebar setiap interval
h = (b - a) / n

# nilai perkiraan awal adalah 0
approximation = 0

# melakukan perulangan sebanyak jumlah interval
for i in range(n):
    # nilai perkiraan akan terus ditambahkan sampai perulangan selesai
    approximation += f(a + i * h)

# nilai perkiraan dikalikan lebar setiap interval
approximation *= h

print(f"Approximation of the integral: {approximation}")

# plotting proses nilai integral dari nilai yang sudah dihitung dengan menggunakan matplotlib
x = np.linspace(a, b, 100)
y = f(x)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.fill_between(x, y, alpha=0.2)
plt.show()
```
