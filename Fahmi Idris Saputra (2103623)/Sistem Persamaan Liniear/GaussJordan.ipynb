```python
# Fahmi Idris Saputra (2103623)
# Pendidikan Ilmu Komputer - A

# Sistem Persamaan Linear GAUSS JORDAN

#Library yang dipakai
import numpy as np
import matplotlib.pyplot as plt
```

```python
# tentukan fungsi gauss jordan untuk mencari SPL Ax = b
# A = sebagai nilai koefisen matriks
# B = sebagai nilai vektor sisi kanan

# fungsi akan mengembalikan nilai X dari sistem persamaan linear Ax = b

def gauss_jordan(A, b):

    # tambahkan matriks koefisien dengan vektor sisi kanan
    # ubah nilai a menjadi float
    A = A.astype(float)

    # perbesar matriks koofesien dengan vektor sisi kanan
    M = np.hstack((A, b.reshape(-1, 1)))
    
    # perhitungan gauss jordan
    for pivot_row in range(M.shape[0]):
        # cari pivot element
        pivot = M[pivot_row][pivot_row]
        
        # normalisasikan baris pada pivot
        M[pivot_row] /= float(pivot)

        # eliminasi nilai kolom pada pivot yang ada di dalam baris yang lain
        for row in range(M.shape[0]):
            if row != pivot_row:
                M[row] -= M[row][pivot_row] * M[pivot_row]
    
    # extract solusi persamaan dari nilai normalisasi sebelumnya dan simpan ke dalam matriks x
    x = M[:, -1]

    return x
```

```python
# membuat matriks a dan b
A = np.array([[2, 1, -1], [-3, -1, 2], [-2, 1, 2]])
b = np.array([8, -11, -3])

# panggil fungsi gauss jordan 
# simpan nilai return nya dalam variabel x
x = gauss_jordan(A, b)

print(f"Solution: {x}")

# plot solusinya
fig, ax = plt.subplots()
x1, x2 = np.meshgrid(np.linspace(-5, 5, 100), np.linspace(-5, 5, 100))
ax.contour(x1, x2, x1 * 2 + x2 - 8, levels=[0])
ax.contour(x1, x2, -x1 * 3 - x2 + 11, levels=[0])
ax.contour(x1, x2, -x1 * 2 + x2 + 3, levels=[0])
ax.scatter(x[0], x[1], marker='x')
plt.show()
```
