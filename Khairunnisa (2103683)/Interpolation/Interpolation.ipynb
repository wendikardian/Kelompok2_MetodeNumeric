```python
# Khairunnisa (2103683)
# Pendidikan Ilmu Komputer - A

# Interpolasi


# Import Libraries
import numpy as np #Untuk inisialisasi nilai matriks dan melakukan proses kalkulasi matematika
import matplotlib.pyplot as plt #Untuk melakukan visualisasi data dan plotting data

# Menentukan titik data matriks yang akan dihitung menggunakan interpolasi
x = np.array([0, 1, 2])
y = np.array([1, 2, 3])
```

```python
# Nilai x interpolasi dengan inisial value = 1.5 yang nilai y nya akan di cari
x_interp = 1.5


# Y akan diisi dengan nilai kalkulasi linear interpolasi berdasarkan rumus yang telah ada
y_interp = (y[1] - y[0]) / (x[1] - x[0]) * (x_interp - x[0]) + y[0]

# Cetak nilai interpolasi
print(f"Interpolated value: {y_interp}")

# Plotting data dari interpolasi yang sudah ditemukan
fig, ax = plt.subplots()
ax.scatter(x, y)
ax.scatter(x_interp, y_interp, marker='x')
plt.show()
```
