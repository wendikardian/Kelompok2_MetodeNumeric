```python
# Khairunnisa (2103683)
# Pendidikan Ilmu Komputer - A

# Nearest Neighbor Interpolation

# Import Libraries
# Untuk mencari nilai x berdasarkan x yang paling dekat
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Mendefinisikan nilai matriks dari data point sebagai data nilai awal yang diketahui
x = np.array([0, 1, 2])
y = np.array([1, 2, 3])

# Mendefinisikan nilai x dari interpolasi yang nantinya akan dicari
x_interp = 1.5

# Mencari nilai index i, dimana nilai datanya paling dekat dengan array x untuk nanti dicocokan ke dalam matrix y
i = np.abs(x - x_interp).argmin()

# Nilai interpolasi y akan diisi dengan nilai yang paling dekat berdasarkan index sebelumnya yang sudah dihitung dan di kalkulasi
# nilai y diambil dari nilai x yang terdekat dari data yang dicari
y_interp = y[i]

# Mencetak nilai interpolasi
print(f"Interpolated value: {y_interp}")

# Plot data
fig, ax = plt.subplots()
ax.scatter(x, y)
ax.scatter(x_interp, y_interp, marker='x')
plt.show()
```
