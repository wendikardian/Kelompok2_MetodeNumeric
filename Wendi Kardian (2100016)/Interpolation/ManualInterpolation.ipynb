```python
# Wendi Kardian (2100016)
# Pendidikan Ilmu Komputer - A

# === INTERPOLATION (CARA MANUAL) =====

# Define data matrix yang nanti akan dicarikan nilai interpolationnya
x = [0, 1, 2]
y = [1, 2, 3]

# Define nilai x  intepolation yang akan dicari
x_interp = 1.5

# Find the indices of the two nearest data points
# Cari nilai dari 2 data point terdekat dari matrix yang sudah di definisikan sebelumnya
# Menggunakan perulangan ... dengan nilai iterasi awal = 0, dan kemudian akan diiterasikan secara terus menerus
i = 0
while x[i] < x_interp:
    i += 1

# jika sudah maka nilai index dikurangi 0, karena index mulai dari 0
i -= 1


# Hitung nilai slope dari 2 data point yang paling dekat
slope = (y[i + 1] - y[i]) / (x[i + 1] - x[i])

# Cari nilai interpolation menggunakan rumus interpolation kemudian tampilkan nilainya
y_interp = slope * (x_interp - x[i]) + y[i]
# print(f"Interpolated value: {y_interp}")
```

```python
# Visualisasi data dari nilai interpolation yang sudah ditemukan

import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.scatter(x, y)
ax.scatter(x_interp, y_interp, marker='x')
plt.show()
```
